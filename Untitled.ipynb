## SomeImportant Graph Names:

### Dist plot : 
Seaborn dist plot  is used to plot a histogram, with some other variations like kdeplot and rugplot.

### Line plot : 
The line plot is one of the most basic plot in seaborn library.  This plot is mainly used to visualize the data in form of some time series, i.e. in continuous manner.

### Lm plot :
The lmplot is another most basic plot. It shows a line representing a linear regression model along with data points on the 2D-space and x and y can be set as the horizontal and vertical labels respectively.

# Line Plot

```python
sns.set_style(style='darkgrid')
sns.lineplot(data=df, x='SepalLengthCm', y='SepalWidthCm')
```

```python
with sns.axes_style(style='whitegrid'):
    sns.lineplot(data=df, x='SepalLengthCm', y='SepalWidthCm', color='green')
```

## Swarm Plot
This is very much similar to stripplot but the only difference is that it does not allow overlapping of markers. It causes jittering in the markers of the plot so that graph can easily be read without information loss as seen in the above plot. 

```python
sns.swarmplot(data=df, x=df.Species, y=df.SepalLengthCm, palette='Accent')
```

## Stripplot
It basically creates a scatter plot based on the category.

```python
sns.stripplot(x=df.Species, y=df.SepalLengthCm)
```

## Countplot
A countplot basically counts the categories and returns a count of their occurrences. It is one of the simplest plots provided by the seaborn library.

```python
sns.countplot(data=df, x=df.Species)
```

```python
with sns.axes_style('whitegrid'):
    plt.figure(figsize=(6,6), dpi=200)
    plt.grid(color='gray', linestyle='--')
    plt.xticks(rotation=90)
    sns.countplot(data=df, y=df.SepalLengthCm, saturation=1)
```

```python
with sns.axes_style('whitegrid'):
    plt.figure(figsize=(6,6), dpi=200)
    plt.grid(color='gray', linestyle='--')
    plt.xticks(rotation=90)
    sns.countplot(data=df, y=df.SepalLengthCm, hue=df.Species, saturation=1)
```

## Box Plot 
It is the visual representation of the depicting groups of numerical data through their quartiles. Boxplot is also used to detect the outlier in the data set.

```python
sns.boxplot(data=df, x=df.Species, y=df.SepalLengthCm, showmeans=True, meanprops={'markersize':10, 'markeredgecolor':'black'})
```

## Violinplot
It is similar to the boxplot except that it provides a higher, more advanced visualization and uses the kernel density estimation to give a better description about the data distribution.

```python
plt.figure(dpi=200)
sns.violinplot(data=df, x=df.Species, y=df.SepalLengthCm)
```

# Scatter Plot

```python
plt.figure(dpi=200)
sns.scatterplot(data=df, x=df.SepalWidthCm, y=df.SepalLengthCm, hue=df.Species, legend=True)
plt.legend(loc='upper right', fontsize=7)
```

### Scatter Plot with lmPlot(Linear Model Regression Plot)

# JointPlot
In data visualization, we often plot the joint behavior of two random variables (bi-variate distribution) or any number of random variables. But if data is too large, overlapping can be an issue. Hence, to distinguish between variables it is useful to have the probability distribution of each variable on the side along with the joint plot. This individual probability distribution of a random variable is referred to as its marginal probability distribution. 

##### Kind:
['scatter', 'hist', 'hex', 'kde', 'reg', 'resid']

```python
sns.jointplot(data=df, x=df.PetalLengthCm, y=df.SepalLengthCm, kind='reg')
sns.jointplot(data=df, x=df.PetalLengthCm, y=df.SepalLengthCm, kind='hist')
sns.jointplot(data=df, x=df.PetalLengthCm, y=df.SepalLengthCm, kind='scatter')
```

# Histogram

```python
plt.figure(dpi=200)
sns.histplot(data=df, x=df.SepalLengthCm, kde=True, bins=30, stat='probability', hue=df.Species)

plt.figure(dpi=200)
sns.histplot(data=df, x=df.SepalLengthCm, kde=True, bins=30, stat='count', color='red')
```

# Dis-Plot

##### Kind:
['hist', 'kde', 'ecdf']

```python
plt.figure(dpi=200)
sns.displot(data=df, x=df.SepalLengthCm, kind='hist', kde=True, bins=30)
```

# Pair - Plot
To plot multiple pairwise bivariate distributions in a dataset, you can use the pairplot() function. This shows the relationship for (n, 2) combination of variable in a DataFrame as a matrix of plots and the diagonal plots are the univariate plots.

```python
sns.pairplot(data=df, hue='Species', height=4)
```

# KDE
Kernel Density Estimate (KDE) Plot and Kdeplot allows us to estimate the probability density function of the continuous or non-parametric from our data set curve in one or more dimensions it means we can create plot a single graph for multiple samples which helps in more efficient data visualization.

```python
sns.kdeplot(x=df.SepalLengthCm, shade=True, color='green')
```

```python
sns.kdeplot(x=df.SepalLengthCm, y=df.PetalLengthCm, shade=True, cbar=True)
```

# HeatMap
Heatmap is defined as a graphical representation of data using colors to visualize the value of the matrix. In this, to represent more common values or higher activities brighter colors basically reddish colors are used and to represent less common or activity values, darker colors are preferred. Heatmap is also defined by the name of the shading matrix. Heatmaps in Seaborn can be plotted by using the seaborn.heatmap() function.

```python
df.corr()
```

```python
sns.heatmap(data=df.corr(), linewidths=2, linecolor='black', cmap='tab20', annot=True)
```
